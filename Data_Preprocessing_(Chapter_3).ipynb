## Handling Missing Value 3 แทนด้วยค่าด้วยค่ากลาง

````python
data_cases[['age']] = data_cases[['age']].fillna(data_cases['age'].mean())
 
````

````python
data_cases.isnull().any()
````

````python
data_cases[['sex']].dropna().mode()
````

````python
data_cases[['sex']].dropna().mode().iloc[0,0]
````

````python
data_cases[['sex']] = data_cases[['sex']].fillna(data_cases[['sex']].dropna().mode().iloc[0,0])
data_cases.isnull().any()
````
